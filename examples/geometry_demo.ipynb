# DeepFlow Geometry Module Demo

This notebook demonstrates the capabilities of the `Geometry` module in the `deepflow` package.
We will cover basic shapes, complex polygons, and shape operations like subtraction.

```python
from deepflow import Geometry
```

## Basic Shapes

### Line
Create a line segment defined by two points.

```python
line = Geometry.line([1,2],[4,5])
line.show()
```

### Rectangle
Create a rectangle defined by x and y ranges.

```python
rectangle = Geometry.rectangle([0,1],[2,1])
rectangle.show()
```

### Polygon
Create a polygon defined by a list of vertices.

```python
shape = Geometry.polygon([0.5,0],[2,0],[2,1],[0,1],[0,0.5],[0.5,0.5])
shape.show()
```

### Circle
Create a circle defined by center (x, y) and radius r.

```python
circle = Geometry.circle(1,0.5,0.2)
X, Y = circle.sampling_lines(100)
circle.show();
```

## Shape Operations

### Subtraction
You can subtract one shape from another. Here we subtract the circle from the polygon defined above.

```python
shape = shape - circle
X,Y = shape.show()
```

## Complex Shapes

### Example 1: "F" Shape
Defining a complex polygon using a list of vertices.

```python
listt = [
    [-5,  6],   # top-left
    [ 5,  6],   # top-right
    [ 5,  4],   # bar bottom-right
    [ 1,  4],   # stem top-right
    [ 1, -6],   # stem bottom-right
    [-1, -6],   # stem bottom-left
    [-1,  4],   # stem top-left
    [-5,  4]    # bar bottom-left
]
complex_shape = Geometry.polygon(*listt)
complex_shape.show()
```

### Example 2: Another Complex Polygon

```python
listt = [
    [0, 0],    # bottom-left
    [0, 10],   # top-left
    [3, 10],   # top of left stroke
    [7, 3],    # diagonal intersection
    [7, 10],   # top of right stroke
    [10, 10],  # top-right
    [10, 0],   # bottom-right
    [7, 0],    # bottom of right stroke
    [3, 7],    # diagonal intersection
    [3, 0],    # bottom of left stroke
]

complex_shape = Geometry.polygon(*listt)
complex_shape.show()
```
