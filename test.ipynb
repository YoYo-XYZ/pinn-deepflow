```python
from deepflow import Geometry
```

```python
line = Geometry.line([1,2],[4,5])
X, Y = line.sampling_line(100)
import matplotlib.pyplot as plt
plt.scatter(X.numpy(), Y.numpy(), s=1)
```

```python
X, Y = Geometry.rectangle([0,0],[2,1]).sampling_area(1000, random=True)
```

```python
shape = Geometry.polygon([0.5,0],[2,0],[2,1],[0,1],[0,0.5],[0.5,0.5])
for i, bound in enumerate(shape):
    X, Y = bound.sampling_line(50)
    plt.scatter(X.numpy(), Y.numpy(), s=1)
    plt.text(X[20],Y[20],i)
print(shape)
```

```python
X,Y = shape.sampling_area([100,100], False)
print(shape.bound_list[0])
plt.scatter(X,Y, s=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
